```python
# Assuming your second dataset is already loaded as stocks_df
# Remove '$' and convert relevant columns to numeric
cols_to_clean = ['Close/Last', 'Open', 'High', 'Low']
for col in cols_to_clean:
    stocks_df[col] = stocks_df[col].replace({'\$': ''}, regex=True).astype(float)

# Convert 'Date' column to datetime format
stocks_df['Date'] = pd.to_datetime(stocks_df['Date'])

# Display prepared stocks_df
stocks_df
```

```python
import pandas as pd

# Convert 'Date' column to datetime
stocks_df['Date'] = pd.to_datetime(stocks_df['Date'], utc=True)

start_date = '2014-11-03' #11/03
end_date = '2020-06-10'
filtered_stocks_df = stocks_df[(stocks_df['Date'] >= start_date) & (stocks_df['Date'] <= end_date)]

filtered_stocks_df
```

```python
# Convert 'date' column to datetime
nvda_df['date'] = pd.to_datetime(nvda_df['date'], utc=True)

start_date = '2014-11-03'
end_date = '2020-06-10'
filtered_nvda_df = nvda_df[(nvda_df['date'] >= start_date) & (nvda_df['date'] <= end_date)]

filtered_nvda_df
```

```python
filtered_nvda_df['date'] = filtered_nvda_df['date'].dt.date

# Group by 'date' and concatenate titles
nvda_df = filtered_nvda_df.groupby('date').agg({
    'title': ' '.join,  # Concatenate all titles for each day
    'stock': 'first'    # Keep stock symbol (assuming it's always NVDA)
}).reset_index()

nvda_df
```

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Load the SpaCy language model
nlp = spacy.load('en_core_web_sm')

# Add the TextBlob sentiment analysis component to the pipeline if not already added
nlp.add_pipe('spacytextblob')

# Function to determine binary sentiment based on polarity
def get_binary_sentiment(text):
    doc = nlp(text)  # Process the text with SpaCy pipeline
    polarity = doc._.blob.polarity  # Get polarity from TextBlob
    # return 1 if polarity > 0 else 0  # Return 1 for positive, 0 for non-positive
    return polarity

# Apply the sentiment analysis model to the 'title' column and create a new 'Sentiment' column
nvda_df['Sentiment'] = nvda_df['title'].apply(get_binary_sentiment)

# Display the updated DataFrame with the new 'Sentiment' column
nvda_df[['date', 'title', 'Sentiment']].head()
```

```python
# Convert date columns to datetime format with UTC
nvda_df['date'] = pd.to_datetime(nvda_df['date'], utc=True)
filtered_stocks_df['Date'] = pd.to_datetime(filtered_stocks_df['Date'], utc=True)

# Merge filtered_stocks_df with nvda_df on the date
final_merged_df = pd.merge(filtered_stocks_df,
                           nvda_df[['date', 'title', 'Sentiment']],
                           left_on='Date',
                           right_on='date',
                           how='left')

# Fill missing values for days without news articles using .loc[]
final_merged_df.loc[:, 'title'] = final_merged_df['title'].fillna("")  # Empty string for missing titles
final_merged_df.loc[:, 'Sentiment'] = final_merged_df['Sentiment'].fillna(0)  # Neutral sentiment (0) for missing articles

# Drop the redundant 'date' column after merge
final_merged_df.drop(columns=['date'], inplace=True)
```

```python
final_merged_df
```

```python
# Create a new binary column 'title_present' indicating if there is a non-empty 'title'
final_merged_df['title_present'] = final_merged_df['title'].str.strip().apply(lambda x: 1 if x != '' else 0)

final_merged_df
```

```python
# Convert 'Date' column to datetime format if not already done
final_merged_df['Date'] = pd.to_datetime(final_merged_df['Date'])

# Sort the DataFrame by Date to ensure correct order for percentage change calculation
final_merged_df = final_merged_df.sort_values(by='Date')

# Calculate the Close-to-Close Percentage Change
# The pct_change() function calculates the percentage change between consecutive rows
final_merged_df['Percentage_Change'] = final_merged_df['Close/Last'].pct_change() * 100

# Display the updated DataFrame with the new column (optional)
final_merged_df[['Date', 'Close/Last', 'Percentage_Change']].head()
```

### Min/Max, Mean,Median, Mode, Standard Deviation

```python
# Add a new column for the length of the 'title' strings
final_merged_df['title_length'] = final_merged_df['title'].fillna('').apply(len)

# Select numeric columns, including the newly added 'title_length'
numeric_columns = final_merged_df.select_dtypes(include='number')

# Calculate statistics
statistics = numeric_columns.describe().T  # Transpose for better readability
statistics['median'] = numeric_columns.median()  # Add median
statistics['mode'] = numeric_columns.mode().iloc[0]  # Add mode (first mode if multiple exist)

# Select relevant columns for display
result = statistics[['min', 'max', 'mean', 'std', 'median', 'mode']]

pd.set_option('display.float_format', '{:,.4f}'.format)

result
```

### Feature Importance: Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Assuming final_merged_df is already loaded and available

# Prepare the data
# Select features (X) and target (Y)
X = final_merged_df[['Volume', 'Open', 'High', 'Low', 'Sentiment', 'Close/Last', 'title_present', 'title_length']]  # Features
Y = final_merged_df['Close/Last']  # Target

# Split the data into training and test sets for better evaluation (best practice)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Feature scaling (optional but recommended for tree-based models)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a Random Forest model
model = RandomForestRegressor(n_estimators=100, random_state=42) # random_state is like a see, and we use 100 trees

# Fit the model to the training data
model.fit(X_train_scaled, Y_train)

# Get feature importance from the model
importances = model.feature_importances_

# Display feature importance
feature_importance_df = pd.DataFrame({'Feature (target = Close/Last)': X.columns, 'Importance': importances})

# Sort features by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

feature_importance_df
```

### Feature Importance: Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor

'''
Why gradient boosting?
* This captures non-linear relationshisp better than other models (e.g. linear regression)
'''

X = final_merged_df[['Volume', 'Open', 'High', 'Low', 'Sentiment', 'Close/Last', 'title_present', 'title_length']]  # Features
Y = final_merged_df['Close/Last']  # Target

# Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Create a Gradient Boosting model
gb_model = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Fit model
gb_model.fit(X_train_scaled, Y_train)

# Get feature importance from the model
gb_importances = gb_model.feature_importances_

# Display feature importance
gb_feature_importance_df = pd.DataFrame({'Feature (target = Close/Last)': X.columns, 'Importance': gb_importances})

# Sort features by importance
gb_feature_importance_df = gb_feature_importance_df.sort_values(by='Importance', ascending=False)

# Reformat Values to 6 decimal points
gb_feature_importance_df['Importance'] = gb_feature_importance_df['Importance'].apply(lambda x: f'{x:.6f}')


gb_feature_importance_df
```
